```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('uber_pool_March.csv')
df2 = pd.read_csv('lyft_line_March.csv')
```

```python
df1 = df1.set_index(pd.DatetimeIndex(df1['date_time']))
df2 = df2.set_index(pd.DatetimeIndex(df1['date_time']))
```

```python
# Merge the two datasets
df = df1.merge(df2, how='inner', on='date_time')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['date_x'], axis=1, inplace=True)
df.drop(['day_x'], axis=1, inplace=True)
df.drop(['time_x'], axis=1, inplace=True)
df.drop(['date_y'], axis=1, inplace=True)
df.drop(['day_y'], axis=1, inplace=True)
df.drop(['time_y'], axis=1, inplace=True)
```

```python
df.columns
```

```python
df.head()
```

Use the same duration of Uber and Lyft

```python
df['average_duration']=(df['lyft_duration'] + df['uber_duration'])/2
```

Adjust the price_per_second using average duration

```python
df['uber_price_per_second']=df['uber_estimate']/df['average_duration']
df['lyft_price_per_second']=df['lyft_estimate']/df['average_duration']
```

```python
df = df.set_index(pd.DatetimeIndex(df1['date_time']))
```

```python
df.drop(['date_time'], axis=1, inplace=True)
```

```python
df.to_csv('uber_lyft_March.csv', sep=',')
```
